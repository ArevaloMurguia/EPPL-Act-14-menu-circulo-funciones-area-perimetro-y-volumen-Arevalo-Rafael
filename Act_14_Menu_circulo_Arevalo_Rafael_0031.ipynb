## 📘 Redaccion del problema:
#### Se requiere desarrollar un programa en Python que permita al usuario calcular diferentes propiedades de un círculo mediante un menú de opciones.

#### El programa deberá mostrar un menú con las siguientes opciones:

1. Calcular el área de un círculo
2. Calcular el perímetro de un círculo
3. Calcular el volumen de una esfera (usando el radio)
4. Salir

#### El usuario podrá elegir una opción y el programa ejecutará la función correspondiente. El sistema se repetirá hasta que el usuario decida salir.

----------

### Variables:
| Variable  | Descripción                         | Tipo   |
| --------- | ----------------------------------- | ------ |
| radio     | Valor del radio del círculo/esfera  | Real   |
| area      | Resultado del área del círculo      | Real   |
| perimetro | Resultado del perímetro del círculo | Real   |
| volumen   | Resultado del volumen de la esfera  | Real   |
| opcion    | Opción seleccionada en el menú      | Entero |

----------

### Formulas:
<img src="https://raw.githubusercontent.com/ArevaloMurguia/EPPL-Act-14-menu-circulo-funciones-area-perimetro-y-volumen-Arevalo-Rafael/refs/heads/main/Circulo%205.jfif" width="300" height="300" alt="Texto descriptivo">

* Área del círculo:

A = π * r²
* Perímetro (circunferencia):

P = 2 * π * r
* Volumen de la esfera:

V = (4/3) * π * r³

```python
# ╔══════════════════════════════════════════════╗
# ║        🔵 CÁLCULOS DE CÍRCULO Y ESFERA 🔵    ║
# ║        Arevalo Rafael NL = 3                 ║
# ╚══════════════════════════════════════════════╝

import math

# ════════════════ MENÚ ════════════════
def menu():
    print("\n╔════════════════════════════════════╗")
    print("║           📌 MENÚ PRINCIPAL        ║")
    print("╠════════════════════════════════════╣")
    print("║ 1️⃣  Área del círculo              ║")
    print("║ 2️⃣  Perímetro del círculo         ║")
    print("║ 3️⃣  Volumen de la esfera          ║")
    print("║ 4️⃣  Salir                         ║")
    print("╚════════════════════════════════════╝")


# ════════════════ ÁREA ════════════════
def areacirculo():
    print("\n╔════════════════════════════════════╗")
    print("║        🟢 ÁREA DEL CÍRCULO         ║")
    print("╚════════════════════════════════════╝")

    radio = float(input("🔹 Ingresa el radio: "))
    area = math.pi * (radio ** 2)

    print("\n╔════════ RESULTADO ════════╗")
    print(f"║ Radio  : {radio:<15} ║")
    print(f"║ Área   : {area:<15.2f} ║")
    print("╚══════════════════════════╝")


# ═══════════════ PERÍMETRO ═══════════════
def perimetrocirculo():
    print("\n╔════════════════════════════════════╗")
    print("║     🔵 PERÍMETRO DEL CÍRCULO       ║")
    print("╚════════════════════════════════════╝")

    radio = float(input("🔹 Ingresa el radio: "))
    perimetro = 2 * math.pi * radio

    print("\n╔════════ RESULTADO ════════╗")
    print(f"║ Radio      : {radio:<10} ║")
    print(f"║ Perímetro  : {perimetro:<10.2f} ║")
    print("╚══════════════════════════╝")


# ═══════════════ VOLUMEN ═══════════════
def volumencirculo():
    print("\n╔════════════════════════════════════╗")
    print("║      🟣 VOLUMEN DE LA ESFERA       ║")
    print("╚════════════════════════════════════╝")

    radio = float(input("🔹 Ingresa el radio: "))
    volumen = (4/3) * math.pi * (radio ** 3)

    print("\n╔════════ RESULTADO ════════╗")
    print(f"║ Radio   : {radio:<12} ║")
    print(f"║ Volumen : {volumen:<12.2f} ║")
    print("╚══════════════════════════╝")


# ═════════════ PROGRAMA PRINCIPAL ═════════════
while True:
    menu()
    opcion = int(input("👉 Selecciona una opción: "))

    if opcion == 1:
        areacirculo()

    elif opcion == 2:
        perimetrocirculo()

    elif opcion == 3:
        volumencirculo()

    elif opcion == 4:
        print("\n🚪 Cerrando programa...")
        break

    else:
        print("\n❌ Opción inválida, intenta de nuevo")

print("\n✔ Programa realizado por Arevalo Rafael NL = 3")
```
